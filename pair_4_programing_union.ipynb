## Ejercicios de Limpieza

1.	Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df_merge.drop("Country", axis = 1, inplace = True)
```

2.	Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

		- No tengan espacios.
		- Estén en minúscula.
		- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
		- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
		- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.
		- La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.
	


```python
nuevas_columnas = {columna:columna.replace(" (%)", "").replace("(Km2)", "").replace("\n", "").replace(":", "").replace("( %)", "").replace(' ', '_').lower() for columna in df_merge.columns}
```

```python
nuevas_columnas
```

```python
df_merge.rename(columns = nuevas_columnas, inplace = True)
```

```python
df_merge.head(1)
```

## Ejercicios de Filtrado
1.	Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.

```python
df_merge.head(2)
```

```python
df_mortality = df_merge[['country', 'infant_mortality']][df_merge ['infant_mortality'].between(40,50, inclusive = 'both')]
df_mortality.reset_index()
```

2.	Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

```python
#primero obtenemos todos los países cuyas tasas de natalidad son mayores o iguales a 20
df_natalidad = df_merge ['country'][df_merge ['birth_rate'] >= 20]
df_natalidad.reset_index()
```

```python
#después, los países cuya esperanza de vida es mayor de 75 años.
df_esp_vida = df_merge ['country'][df_merge ['life_expectancy'] >75]
df_esp_vida.reset_index()
```

```python
#unimos las dos condiciones y lo igualamos a una variable para que nos lo muestre en forma de dataframe
df_vida = df_merge [(df_merge ['birth_rate'] >= 20) & (df_merge ['life_expectancy'] >75)]
df_vida
```
