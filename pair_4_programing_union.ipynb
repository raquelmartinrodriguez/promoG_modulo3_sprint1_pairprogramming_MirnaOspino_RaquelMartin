3.	Utiliza el método de unión de Pandas que consideres más adecuado para combinar los datos de ambos años en un solo DataFrame.

```python
df_merge = df1.merge(df2, left_on= "Country", right_on= "country", how= "inner")
```

4.	Explica por qué elegiste ese método de unión y cómo se llevaron a cabo los pasos anteriores:  Encontramos la columna común "country" y decidimos hacer un merge con inner a partir de esta columna. De esta manera unimos los datos de cada pais en una sola fila.

1.	Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df_merge.drop("Country", axis = 1, inplace = True)
```

2.	Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
o	No tengan espacios.
o	Estén en minúscula.
o	No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
o	Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
o	Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.
	La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.
	


```python
nuevas_columnas = {columna:columna.replace(" (%)", "").replace("(Km2)", "").replace("\n", "").replace(":", "").replace("( %)", "").replace(' ', '_').lower() for columna in df_merge.columns}
```

```python
nuevas_columnas
```

```python
df_merge.rename(columns = nuevas_columnas, inplace = True)
```

```python
df_merge.head(1)
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_merge['coordinates'].str.split(",", expand = True)
```

```python
df_merge[["latitud", "longitud"]] = df_merge["coordinates"].str.split(",", expand = True).get([0,1])
```

```python
df_merge[["latitud", "longitud"]]
```

```python
df_merge.drop("coordinates", axis = 1, inplace = True)
```

```python
df_merge.columns
```

4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.
	Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".
	Guarda el DataFrame para usarlo en el pairprogramming de mañana

```python
columns_to_clean = ["unemployment_rate", "total_tax_rate", "tax_revenue", "population_labor_force_participation", "out_of_pocket_health_expenditure", "gross_tertiary_education_enrollment", "gross_primary_education_enrollment", "forested_area", "cpi_change", "agricultural_land"]

for column in columns_to_clean:
    df_merge[column] = df_merge[column].str.replace("%", "")
```

```python
df_merge[columns_to_clean]
```

```python
df_merge["unemployment_rate"]
```

5.	Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columns_to_clean2 = ["gasoline_price", "gdp", "minimum_wage"]
for column in columns_to_clean2:
    df_merge[column] = df_merge[column].str.replace("$", "")
```
