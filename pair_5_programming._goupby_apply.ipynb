1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np


# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# cargamos el dataframe creado en la lección anterior
df = pd.read_csv("world_data_total.csv", index_col = 0)
df.head(6)
```

```python
df [['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitud', 'longitud', 'density(p/km2)', 'cpi']]
```

```python
def cambiar_tipo (num):
    try:
        nuevo_tipo = float(num.replace (',', ''))
        return nuevo_tipo

    except: 
        return np.nan
```

```python
#empezamos por las columnas en las cuales tenemos que eliminar las comas para que sean de tipo numérico
columnas = ['armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'density(p/km2)', 'cpi']
```

```python
for columna in columnas:
    df [columna] = df [columna].apply (cambiar_tipo)
```

```python
df.head()
```

```python
#a las siguientes columnas hay que realizarles más cambios para poder trabajar con ellas
col_lat_long = ['latitud', 'longitud', 'land_area']
```

```python
def cambiar_tipo2 (num):
    try:
        nuevo_tipo = num.replace ('(', ''). replace (')', ''). replace (' ', ''). replace (',', ''). replace ('.', '').replace ('"', '')
        
    
        return nuevo_tipo

    except: 
        return np.nan
```

```python
for columna in col_lat_long:
    df [columna] = df [columna].apply (cambiar_tipo2)
```

```python
df['latitud'].unique()
#todavía no podemos trabajar con ellas. Debemos convertirlas a integer
```

```python
# Quitar las comillas de los valores en la columna latitud
df['latitud'] = df['latitud'].str.replace("'", "")

# Convertir los valores de la columna latitud a tipo integer
df['latitud'] = df['latitud'].astype('Int64', errors='ignore')

# Quitar las comillas de los valores en la columna longitud
df['longitud'] = df['longitud'].str.replace("'", "")

# Convertir los valores de la columna longitud a tipo integer
df['longitud'] = df['longitud'].astype('Int64', errors='ignore')

# Quitar las comillas de los valores en la columna longitud
df['land_area'] = df['land_area'].str.replace("'", "").replace(',', '')

# Convertir los valores de la columna longitud a tipo integer
df['land_area'] = df['land_area'].astype('Int64', errors='ignore')
```

5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:
¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:

Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".

Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".

Ahora, divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.

```python
# Primero, calculamos la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna columna "total_tax_rate".
df_tasa = df.groupby ('continent')["total_tax_rate"].sum().reset_index()
df_tasa
```

```python
#Luego, calculamos la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".
df_densidad = df.groupby ('continent')['density(p/km2)'].sum().reset_index()
df_densidad
```

```python
# Combinar los dos DataFrames en uno solo utilizando merge en la columna 'continent'
df_merged = pd.merge(df_tasa, df_densidad, on='continent')


# Calculamos la tasa de impuestos total por persona por kilómetro cuadrado y lo almacenamos en una nueva columna llamada 'tax_rate_per_person_per_km2'
df_merged['tax_rate_per_person_per_km2'] = df_merged['total_tax_rate'] / df_merged['density(p/km2)']
df_merged
```

```python
max_tax_rate_per_person_per_km2 =df_merged['tax_rate_per_person_per_km2'].max()
```

```python
df_maximo= df_merged['continent'][df_merged['tax_rate_per_person_per_km2'] == max_tax_rate_per_person_per_km2]
df_maximo.reset_index()
```
